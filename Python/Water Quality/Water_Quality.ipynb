# Water Potability : Scatter with all factors

```python
plt.figure(figsize=(15,15))
for n in range(9):
  plt.subplot(3, 3, n + 1)
  plt.scatter(df.iloc[:, n:n + 1], df.iloc[:, -1:])
  plt.title(df.columns[n])
plt.show()
```

No visual pattern for water potability with any factor

# Correlation Matrix

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(), cmap='vlag_r', center=0, vmin=-1, vmax=1, annot=True)
```

Negligible Correlation between all factors and water potability
